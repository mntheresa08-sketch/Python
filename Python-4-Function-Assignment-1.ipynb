```python
# Create a function that checks whether the given number is Odd or Even
#OddEven.OddEven()
#Enter a number: 52452
#52452 is Even number
```

```python
class OddEven():
    def OddEven():
        num=int(input("Enter your number here:"))
        if(num%2 == 0):
            return (f"Enter a number:{num}\n"
                    f"{num} is Even number")
        else:
            return (f"Enter a number:{num}\n"
                    f"{num} is Odd number")
print (OddEven.OddEven())
```

```python
## # Create a function that tells elegibility of marriage for male and female according to their age limit like 21 for male and 18 for female
#ElegiblityForMarriage.Elegible()
#Your Gender:Male
#Your Age:20
#NOT ELIGIBLE
```

```python
class ElegiblityForMarriage():

    def Elegible():
        gender = input("Enter your gender (male/female): ")
        age = int(input("Enter your age: "))

        if gender.lower() == "male" and age >= 21:
            status = "Eligible"
        elif gender.lower() == "female" and age >= 18:
            status = "Eligible"
        elif gender.lower() in ["male", "female"]:
            status = "Not Eligible"
        else:
            return "Invalid input."

        return f"Gender: {gender} \n| Age: {age} \n| Status: {status} for marriage"

print(ElegiblityForMarriage.Elegible())
```
